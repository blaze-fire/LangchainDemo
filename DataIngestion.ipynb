# Embeddings

## Ollama Embeddings

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings = (
    OllamaEmbeddings(model="gemma2:2b")  #by default uses llama2
)
```

```python
r1 = embeddings.embed_documents(
    [
        "Alpha is first",
        "Beta is the second"
    ]
)
```

```python
r1[0]
```

```python
embeddings.embed_query("Who is the second ")
```

## Other Embedding Models
